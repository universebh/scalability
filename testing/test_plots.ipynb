# Error Bar

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Throughput vs. number of cores (2 curves)

```python
throughput_pr = np.array(
    [[56.1906, 52.1527, 67.8798, 71.0475, 85.6392],
    [86.04, 118.712, 82.9395, 110.439, 98.1843],
    [133.721, 109.311, 117.982, 124.302, 126.273]])
throughput_tp = np.array(
    [[7.50762, 7.47231, 7.52384, 7.77634, 7.50509],
    [7.62342, 7.62838, 7.66904, 7.51487, 7.68843],
    [7.61735, 7.81952, 7.88469, 7.66441, 7.9092]])

core = np.array([1., 2., 4.])

plt.figure(figsize=(10, 6))
plt.errorbar(core, np.average(throughput_pr, axis=1), yerr=np.std(throughput_pr, axis=1), fmt='.--', label='thread per request')
plt.errorbar(core, np.average(throughput_tp, axis=1), yerr=np.std(throughput_tp, axis=1), fmt='.--', label='thread pool')
plt.legend()
plt.xlabel('number of cores')
plt.ylabel('throughput (req/s)')
plt.title('Throughput vs. number of cores')
plt.show()
```
